# Mapeamento

```python
def mapeamento(df):
    df_map = pd.DataFrame(df)
    df_map
    sexo_map = {
        1: 'Feminino',
        2: 'Masculino'
    }   

    
    FSO_map = {
        1: 'Sim',
        2: 'Não',
    }
    CFF_map = {
        1: 'Sim',
        2: 'Não',
    }
    FCV_map = {
        1: 'Raramente',
        2: 'Às vezes',
        3: 'Sempre',
    }
    NRPD_map = {
        1: '1-2',
        2: '3',
        3: '3+',
    }
    IAER_map = {
        1: 'Raramente',
        2: 'Às vezes',
        3: 'Normalmente',
        4: 'Sempre'
    }
    CIC_map = {
        1: 'Sim',
        2: 'Não',
    }
    Fumante_map = {
        1: 'Sim',
        2: 'Não',
    }
    IDL_map = {
        1: 'Menor que um litro',
        2: 'Entre 1 e 2 litros',
        3: 'Superior a 2 litros'
    }
    EF_map = {
        1: 'Nenhuma atividade física',
        2: 'De 1 a 2 dias',
        3: 'De 3 a 4 dias',
        4: 'De 5 a 6 dias',
        5: '6+ dias'
    }
    HDT_map = {
        1: 'Entre 0 e 2 horas',
        2: 'Entre 3 e 5 horas',
        3: 'Mais de cinco horas'
    }
    TTU_map = {
        1: 'Automóvel',
        2: 'Motocicleta',
        3: 'Bicicleta',
        4: 'Transporte público',
        5: 'Caminhada'
    }
    classificacao_map = {
        1: 'Abaixo do peso',
        2: 'Normal',
        3: 'Sobrepeso',
        4: 'Obesidade',
    }
    
    df_map['Classificação'] = df_map['Classificação'].map(classificacao_map)
    df_map['Tipo de Transporte Utilizado'] = df_map['Tipo de Transporte Utilizado'].map(TTU_map)
    df_map['Horário Dedicado à Tecnologia'] = df_map['Horário Dedicado à Tecnologia'].map(HDT_map)
    df_map['Exercício Físico'] = df_map['Exercício Físico'].map(EF_map)
    df_map['Ingestão Diária de Líquidos'] = df_map['Ingestão Diária de Líquidos'].map(IDL_map)
    df_map['Fumante'] = df_map['Fumante'].map(Fumante_map)
    df_map['Cálculo da Ingestão Calórica'] = df_map['Cálculo da Ingestão Calórica'].map(CIC_map)    
    df_map['Ingestão de Alimentos Entre Refeições'] = df_map['Ingestão de Alimentos Entre Refeições'].map(IAER_map)
    df_map['Número de Refeições Principais Diárias'] = df_map['Número de Refeições Principais Diárias'].map(NRPD_map)    
    df_map['Frequência de Consumo de Vegetais'] = df_map['Frequência de Consumo de Vegetais'].map(FCV_map)
    df_map['Consumo de Fast Food'] = df_map['Consumo de Fast Food'].map(CFF_map)    
    df_map['Sexo'] = df_map['Sexo'].map(sexo_map)    
    df_map['Famílias com Sobrepeso/Obesidade'] = df_map['Famílias com Sobrepeso/Obesidade'].map(FSO_map)
    return df_map
```

# Dividir Dados

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, random_state = 42,test_size = 0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Treinamento

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
test_model = [LogisticRegression(random_state =42),
              DecisionTreeClassifier(random_state =42),
              RandomForestClassifier(random_state =42), 
              ExtraTreesClassifier(random_state =42), 
              GradientBoostingClassifier(random_state =42), 
              AdaBoostClassifier(random_state =42),
              XGBClassifier(random_state=42)]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
model = ['Regressão Logística',
         'Classificador de árvore de decisão',
         'Classificador Florestal Aleatório',
         'Classificador de árvores extras',
         'Classificador de aumento de gradiente',
         'Classificador Ada Boost',
         'XGB']
result = pd.DataFrame(columns=['Modelo','Exatidão','Precisão','Revocação ','F1'])
for i, x in enumerate(test_model):
    train_model = x.fit(X_train,y_train)
    test_model = x.predict(X_test)
    accuracy_scores = accuracy_score(y_test,test_model)
    precision_scores = precision_score(y_test,test_model, average = 'macro')
    recall_scores = recall_score(y_test,test_model,average = 'macro')
    f1_scores = f1_score(y_test,test_model,average = 'macro')
    results = pd.DataFrame({'Modelo':[model[i]],
                           'Exatidão':[accuracy_scores],
                           'Precisão':[precision_scores],
                           'Revocação ':[recall_scores],
                           'F1':[f1_scores]})
    result = pd.concat([result,results],ignore_index = True)
result
```

```python
plt.figure(figsize=(20,10))
for i, a in enumerate(['Exatidão','Precisão','Relembrar','F1']):
    plt.subplot(2,2,i+1)
    sns.barplot(data = result.sort_values(by=a, ascending = False), x= a, y = 'Modelo', palette= 'magma')
plt.tight_layout()
plt.show()
```


Classificador de árvores extra é o melhor modelo para este caso, o desempenho do modelo tem 93,3% de exatidão, 93,5% de precisão, 93,5% de relembrar e 93,4% de pontuação F1.

# Uma Sociedade em Movimento

Uma sociedade em constante movimento. As pessoas com diferentes hábitos, rotinas e prioridades. Nossos gráficos nos convidam a explorar como as escolhas que fazemos no dia a dia, desde a forma como nos alimentamos até os meios de transporte que utilizamos, podem influenciar diretamente nossa saúde.

### A Dieta e a Contagem de Calorias

Nos gráficos mostra que aqueles que se preocupam em contar as calorias que consomem tendem a ter um peso mais saudável. Isso sugere que a conscientização sobre o que comemos é um passo fundamental para controlar o peso. No entanto, a contagem de calorias por si só não garante o sucesso. A qualidade dos alimentos também desempenha um papel crucial.

### O Sedentarismo Moderno

Nos gráficos é revelado a sombra do sedentarismo. Aqueles que passam longas horas em frente às telas, seja trabalhando ou por lazer, tendem a ganhar mais peso. A falta de atividade física, combinada com hábitos alimentares pouco saudáveis, cria um cenário propício para o sobrepeso e a obesidade.

### O Poder da Atividade Física

Mas nem tudo está perdido. Em algums gráfico nos mostra que a prática regular de atividade física pode reverter esse quadro. Aqueles que se dedicam a exercícios físicos regularmente apresentam um peso mais saudável. Seja na academia, em atividades ao ar livre ou simplesmente caminhando, o movimento é um aliado poderoso na luta contra o excesso de peso.

### Escolhas Diárias, Impactos duradouros

Nos gráficos nos leva a refletir sobre as escolhas que fazemos no nosso dia a dia. A forma como nos locomovemos, por exemplo, tem um impacto significativo em nossa saúde. Aqueles que utilizam meios de transporte ativos, como a bicicleta e a caminhada, tendem a ser mais magros.

### Um Mosaico Complexo

Ao analisar todos os gráficos juntos, percebemos que a relação entre estilo de vida e peso é complexa e multifacetada. Não existe uma única fórmula mágica para alcançar o peso ideal. A saúde é resultado de um conjunto de hábitos e escolhas que fazemos ao longo da vida.

### Conclusão




A jornada que fizemos através desses gráficos nos mostra que pequenas mudanças no nosso dia a dia podem ter um grande impacto na nossa saúde. Ao adotar hábitos mais saudáveis, como uma alimentação equilibrada, a prática regular de atividade física e a redução do tempo sedentário, podemos melhorar nossa qualidade de vida e prevenir o sobrepeso e a obesidade.
