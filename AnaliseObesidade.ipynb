# Mapeamento

Sexo

```python
sexo_map = {
    1: 'Feminino',
    2: 'Masculino',
}
df['Sexo'] = df['Sexo'].map(sexo_map)
```

Consumo de Fast Food

```python
CFF_map = {
    1: 'Sim',
    2: 'Não',
}
df['Consumo de Fast Food'] = df['Consumo de Fast Food'].map(CFF_map)
```

Frequência de Consumo de Vegetais

```python
FCV_map = {
    1: 'Raramente',
    2: 'Às vezes',
    3: 'Sempre',
}
df['Frequência de Consumo de Vegetais'] = df['Frequência de Consumo de Vegetais'].map(FCV_map)
```

Número de Refeições Principais Diárias

```python
NRPD_map = {
    1: '1-2',
    2: '3',
    3: '3+',
}
df['Número de Refeições Principais Diárias'] = df['Número de Refeições Principais Diárias'].map(NRPD_map)
```

Ingestão de Alimentos Entre Refeições

```python
IAER_map = {
    1: 'Raramente',
    2: 'Às vezes',
    3: 'Normalmente',
    4: 'Sempre'
}
df['Ingestão de Alimentos Entre Refeições'] = df['Ingestão de Alimentos Entre Refeições'].map(IAER_map)
```

Cálculo da Ingestão Calórica

```python
CIC_map = {
    1: 'Sim',
    2: 'Não',
}
df['Cálculo da Ingestão Calórica'] = df['Cálculo da Ingestão Calórica'].map(CIC_map)
```

Fumante

```python
Fumante_map = {
    1: 'Sim',
    2: 'Não',
}
df['Fumante'] = df['Fumante'].map(Fumante_map)
```

Ingestão Diária de Líquidos

```python
IDL_map = {
    1: 'Menor que um litro',
    2: 'Entre 1 e 2 litros',
    3: 'Superior a 2 litros'
}
df['Ingestão Diária de Líquidos'] = df['Ingestão Diária de Líquidos'].map(IDL_map)
```

Exercício Físico

```python
EF_map = {
    1: 'Nenhuma atividade física',
    2: 'De 1 a 2 dias',
    3: 'De 3 a 4 dias',
    4: 'De 5 a 6 dias',
    5: '6+ dias'
}

# Replacing numerical values with text descriptions
df['Exercício Físico'] = df['Exercício Físico'].map(EF_map)
```

* **Horário Dedicado à Tecnologia:**
  * Entre 0 e 2 horas (382)
  * Entre 3 e 5 horas (826)
  * Mais de cinco horas (402)

```python
HDT_map = {
    1: 'Entre 0 e 2 horas',
    2: 'Entre 3 e 5 horas',
    3: 'Mais de cinco horas'
}
df['Horário Dedicado à Tecnologia'] = df['Horário Dedicado à Tecnologia'].map(HDT_map)
```

Tipo de Veiculo

```python
TTU_map = {
    1: 'Automóvel',
    2: 'Motocicleta',
    3: 'Bicicleta',
    4: 'Transporte público',
    5: 'Caminhada'
}

# Replacing numerical values with text descriptions
df['Tipo de Transporte Utilizado'] = df['Tipo de Transporte Utilizado'].map(TTU_map)
```

Classificação

```python
classificacao_map = {
    1: 'Abaixo do peso',
    2: 'Normal',
    3: 'Sobrepeso',
    4: 'Obesidade',
}
df['Classificação'] = df['Classificação'].map(classificacao_map)
```

# Analises

```python
plt.figure(figsize=(12, 6)) 
sns.countplot(data=df, x='Classificação')
plt.xlabel('Classificação de Obesidade', fontsize=12)
plt.ylabel('Quantidade de Indivíduos', fontsize=12) 
plt.title('Distribuição das Classificações de Obesidade', fontsize=14) 
plt.xticks(ha='right') # Rotate x-axis labels for better readability
plt.show()
```

```python

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(data = df, x = 'Idade', kde= True, color = 'purple' )
plt.title('Distribuição de Idade', fontsize=14)
plt.ylabel('Quantidade de Indivíduos', fontsize=12) 
plt.subplot(1,2,2)
sns.histplot(data = df, x = 'Altura', kde= True, color = 'purple' )
plt.ylabel('Quantidade de Indivíduos', fontsize=12) 
plt.title('Distribuição de Altura', fontsize=14)
 
```
